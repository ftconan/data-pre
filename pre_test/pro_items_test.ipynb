```python
size_order = sorted(list({(item['size_name'], item['size_order']) for item in data}), key=lambda x: x[1])
size_order = [i[0] for i in size_order]
size_order
```

```python
import json
json_df = pd.read_json(json.dumps(data))
```

```python
new_df1 = pd.pivot_table(json_df, 
                        values=['produce_quantity'],
                        index=['color_code', 'warehouse', 'set_code'],
                        columns=['size_name'],
                        fill_value=0,
                        aggfunc=np.sum,
                        dropna=True)
```

```python
new_df1
```

```python
new_df1.columns
df1 = new_df1.reindex(columns=size_order, level='size_name')
df1
```

```python
# pd.concat([new_df1, new_df2], join="inner")
# new_df = new_df1.append(df2)
new_table = pd.concat([
                d.append(d.sum().rename(('', '加裁', '')))
                for k, d in df1.groupby(level=['color_code', 'warehouse'])
            ])
# new_table = pd.concat([
#                 d.append(x)
#                 for k, d in df1.groupby(level='color_code')
#             ])
# df1.dtypes
# df1.append(x)
```

```python
new_table.loc[('', '加裁', '')] = new_table.loc[('', '加裁', '')] * 0.03
new_table.loc[('', '加裁', '')]
```

```python
new_table.loc[:] = round(new_table.loc[:])
new_table.loc[:] = new_table.loc[:].astype('int64')
new_table
```

```python
new_table = new_table.append(new_table.sum().rename(('合计', '', '')))
new_table[('produce_quantity', '合计')] = new_table.sum(axis=1)
new_table
```

```python
new_table1 = new_table.copy(deep=True)
new_table1.columns = new_table1.columns.droplevel(0)
new_table1
```
